```python
from magma_multigas import MultiGas, Diagnose
```

## Defining dataset

```python
# 
two_seconds = 'D:\\Data\\TANG_RTU_ChemData_Sec2.dat'

# CO2 - SO2 - H2S Concentration, CO2/H2S Ratio, H2O/CO2 Ratio, H2S/SO2 Ratio, CO2/SO2 Ratio
# CO2/Stotal Ratio, Sulfur Speciation (%) = Avg_SO2_proportion, Avg_H2S_proportion
six_hours = 'D:\\Data\\TANG_RTU_Data_6Hr.dat'

# Air Temperature, Air Humidity, Battery Voltage, Wind Speed, Wind Direction
one_minute = 'D:\\Data\\TANG_RTU_Wx_Min1.dat'

# CO2 - Zero, H2S - Zero, SO2 - Zero
zero = 'D:\\Data\\TANG_RTU_Zero_Data.dat'
```

```python
multigas = MultiGas(
    two_seconds=two_seconds,
    six_hours=six_hours,
    one_minute=one_minute,
    zero=zero,
    overwrite=True,
)
```

### Extract daily of multigas data

```python
availability = multigas.extract_daily()
```

```python
availability.keys()
```

```python
availability['one_minute']
```

### Selecting ALL dataset between two dates

```python
multigas = multigas.where_date_between(start_date='2024-05-17', end_date='2024-07-24')
```

### Select `two_seconds` data

```python
two_seconds_data = multigas.select('two_seconds').get()
```

```python
two_seconds_data.metadata['station']
```

```python
two_seconds_data.df
```

### Check availabile columns in `two_seconds` data

```python
two_seconds_data.columns    
```

### Query dataset 

```python
filtered_two_seconds = (two_seconds_data
                        .select_columns(column_names=['H2O','CO2_lowpass','SO2','H2S','S_total'])
                        .where_date_between(start_date='2024-05-17', end_date='2024-07-24')
                        .where('Status_Flag', '==', 1)
                       )
```

### See query result

```python
filtered_two_seconds.get()
```

## Select `six_hours` dataset

```python
six_hours_data = multigas.select('six_hours').get()
```

### Select `six_hours` dataset between two dates

```python
filtered_six_hours = six_hours_data.where_date_between(start_date='2024-05-17', end_date='2024-07-24')
```

#### Initiate plot for `six_hours` dataset

```python
plot_six_hours = filtered_six_hours.plot()
```

#### Plot CO2, SO2 and H2S from `six_hours` dataset 

```python
plot_six_hours.plot_co2_so2_h2s()
```

### Refresh `six_hours` dataset the select dataset between dates

```python
filtered_six_hours.refresh()
filtered_six_hours = six_hours_data.where_date_between(start_date='2024-05-17', end_date='2024-07-24')
```

```python
filtered_six_hours.df
```

#### Filtering `six_hours` dataset with value of `Avg_CO2_lowpass` must be greater than equal 250

```python
filtered_six_hours = filtered_six_hours.where('Avg_CO2_lowpass', '>=', 250)
```

#### Plot CO2, SO2 and H2S from `six_hours` dataset after removing anomaly

```python
filtered_six_hours.plot().plot_co2_so2_h2s(
    y_left_min=400, y_left_max=460,
    y_right_min=-0.5, y_right_max=5)
```

#### Plot as individual

```python
filtered_six_hours.plot().plot_co2_so2_h2s(plot_as_individual=True)
```

#### Plot gas ratio

```python
filtered_six_hours.plot(
    height=3
).plot_gas_ratio(
    plot_type='plot',
    space_between_plot=0.2,
    y_max_multiplier=1.1,
    plot_regression=False
)
```

#### Check `six_hours` columns

```python
filtered_six_hours.df.columns
```

```python
six_hours_plot = filtered_six_hours.plot(
    width=8, height=8, margins=0.08, style="white")
```

```python
six_hours_plot.df
```

### Plot `correlogram`

```python
six_hours_plot.plot_correlogram(
    columns=['Avg_H2O', 'Avg_CO2_lowpass', 'Avg_SO2', 'Avg_H2S'],
    plot_regression=False, kind="scatter")
```

### Plot `density`

```python
six_hours_plot.plot_density(
    x='Avg_CO2_lowpass', 
    y='Avg_H2S', 
    kind='scatter',
    kwargs={
        'color': 'm', 
        's' : 100, # Comment this if you use 'kde'
        'marginal_ticks' : True,
        'alpha' : .5}
)
```

### Plot selected column(s)

```python
filtered_six_hours.plot(height=3).plot_columns(
    columns=['Avg_CO2_H2S_ratio', 'Avg_H2O_CO2_ratio', 'Avg_H2S_SO2_ratio',
                   'Avg_CO2_S_tot_ratio'],
    plot_regression=True,
    space_between_plot=0.2,
    disable_regression_for=['Avg_Wind_Direction']
)
```

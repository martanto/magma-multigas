```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize
from matplotlib import colormaps
import pandas as pd
from datetime import datetime, timedelta
from magma_multigas import MultiGasData
```

```python
list(colormaps)
```

```python
csv = r"D:\Projects\magma-multigas\output\normalize\tangkuban_parahu\TANG_RTU_Data_6Hr.dat"
```

```python
multigas = MultiGasData('six_hours',csv)
```

```python
multigas = multigas.where_date_between(start_date='2024-05-17', end_date='2025-03-03')
```

```python
df = multigas.df
```

```python
df.head()
```

```python
wind_speed = df['Avg_Wind_Speed']
wind_direction = df['Avg_Wind_Direction']
u = -np.sin(np.radians(wind_direction))
v = -np.cos(np.radians(wind_direction))
co2_concentration = df['Avg_CO2_S_tot_ratio']
```

```python
fig, ax = plt.subplots(figsize=(12, 4))

# Create a colormap based on wind speed
norm = Normalize(vmin=np.min(wind_speed), vmax=np.max(wind_speed))
cmap = colormaps['rainbow']

# Plot the quiver (arrows) with wind speed as color
# Now using CO2 concentration for Y-axis
q = ax.quiver(df.index, co2_concentration, 
              u, v, wind_speed, cmap=cmap, norm=norm,
              scale=50, width=0.003, pivot='mid')

# Format x-axis as dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=14))
plt.xticks(rotation=30)

# Add labels and title
ax.set_xlabel('Time')
ax.set_ylabel('CO2/S Total Ratio')
ax.set_title('Tangkuban Parahu - CO2/S Total Ratio\n 2024-05-17 to 2025-03-03')
ax.set_ylim([-5, 125])

# Add color bar for wind speed
cbar = fig.colorbar(ScalarMappable(norm=norm, cmap=cmap), ax=ax)
cbar.set_label('Wind Speed (m/s)')

# Add arrow scale reference
ax.quiverkey(q, 0.7, 0.9, 1, r'Wind Direction', labelpos='E', coordinates='figure')

# Add grid lines
ax.grid(True, linestyle='--', alpha=0.3)

# Adjust layout
plt.tight_layout()
plt.show()
```
